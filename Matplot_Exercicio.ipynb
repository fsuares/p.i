```python
import matplotlib.pyplot as plt
import numpy as np

mes = np.arange(12) + 1 # meses de 1 a 12

# quantidades de produtos vendidos por mês
creme_facial = np.array([ 2500, 2630, 2140, 3400, 3600, 2760, 2980, 3700, 3540, 1990, 2340, 2900 ])
limpeza_facial = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])
pasta_dentaria = np.array([ 5200, 5100, 4550, 5870, 4560, 4890, 4780, 5860, 6100, 8300, 7300, 7400 ])
sabonete = np.array([ 9200, 6100, 9550, 8870, 7760, 7490, 8980, 9960, 8100, 10300, 13300, 14400 ])
shampoo = np.array([ 1200, 2100, 3550, 1870, 1560, 1890, 1780, 2860, 2100, 2300, 2400, 1800 ])
hidratante = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])
```

# Gráfico 1 - Total de produtos Vendidos por mês - Linha

```python
tabela_final = np.array([creme_facial, limpeza_facial, pasta_dentaria, sabonete, shampoo, hidratante])
tabela_final.shape = (6, 12)
total_mes = []
for i in range(12):
    soma = 0
    for j in range(6):
        soma = soma + tabela_final[j, i]
    total_mes.append(soma)

np.array(total_mes)

x = mes
y = total_mes
fig, ax = plt.subplots()

ax.plot(x, y)
fig.set_size_inches(10,7)
fig.show()
```

# Gráfico 2 - Gráfico com todos os produtos vendidos por mês - Linha

```python
x = mes

fig, ax = plt.subplots()
ax.plot(x, creme_facial, label='Creme facial')
ax.plot(x, limpeza_facial, label='Limpeza facial')
ax.plot(x, pasta_dentaria, label='Pasta de dente')
ax.plot(x, sabonete, label='Sabonete')
ax.plot(x, shampoo, label='Shampoo')
ax.plot(x, hidratante, label='Hidratante')

ax.legend()

fig.set_size_inches(10,7)
fig.show()
```

# Gráfico 3 - Comparativo de Creme Facial com Limpeza Facial por mês - Barras


```python
fig, ax = plt.subplots()
barra = 0.4
ax.bar(mes - barra/2, creme_facial, width=barra, label='Creme')
ax.bar(mes + barra/2, limpeza_facial, width=barra, label='Limpeza')

ax.set(xlim=(0,13))
ax.legend()
fig.set_size_inches(10,7)
plt.show()
fig.show()
```

# Gráfico 5 - Pizza. % da quantidade produtos vendidos no ano em cada produto

```python
fig, ax = plt.subplots()
fig.set_size_inches(10,7)

ano_creme = 0
ano_limpeza = 0
ano_pasta = 0
ano_sabonete = 0
ano_shampoo = 0
ano_hidratante = 0

for i in range (12):
  ano_creme = ano_creme + creme_facial[i]
  ano_limpeza = ano_limpeza + limpeza_facial[i]
  ano_pasta = ano_pasta + pasta_dentaria[i]
  ano_sabonete = ano_sabonete + sabonete[i]
  ano_shampoo = ano_shampoo + shampoo[i]
  ano_hidratante = ano_hidratante + hidratante[i]

total = ano_creme + ano_limpeza + ano_pasta + ano_sabonete + ano_shampoo + ano_hidratante

porc_creme = (ano_creme/total)*100
porc_limpeza = (ano_limpeza/total)*100
porc_pasta = (ano_pasta/total)*100
porc_sabonete = (ano_sabonete/total)*100
porc_shampoo = (ano_shampoo/total)*100
porc_hidratante = (ano_hidratante/total)*100


y = np.array([porc_creme, porc_limpeza, porc_pasta, porc_sabonete, porc_shampoo, porc_hidratante])
legend = np.array(['Creme', 'Limpeza', 'Pasta', 'Sabonete', 'Shampoo', 'Hidratante'])

ax.pie(y, labels=legend, shadow=True, autopct='%1.1f%%')
fig.show()
```
